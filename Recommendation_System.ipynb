## Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

##Importing Data

```python
df = pd.read_excel('/content/movies.xlsx')
df.head()
```

```python
movie_titles = pd.read_csv("/content/Movie_Id_Titles.csv")
movie_titles.head()
```

```python
#We can merge them together
df = pd.merge(df,movie_titles,on='movie_id')
df.head()
```

## EDA

```python
#Let's create a ratings dataframe with average rating and number of ratings
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
#Now set the number of ratings column
ratings['ratings_count'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
#Now a few histograms to have a general idea of what the data looks like
ratings['ratings_count'].hist(bins=70)
```

```python
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x='rating',y='ratings_count',data=ratings,alpha=0.5)
```

## Recommending Similar Movies
Now let's create a matrix/pivot table that has the user ids on one access and the movie title on another axis. Each cell will then consist of the rating the user gave to that movie. Note there will be a lot of NaN values, because most people have not seen most of the movies

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
len(moviemat)
```

```python
#Most rated movie
ratings.sort_values('ratings_count',ascending=False).head(10)
```

```python
#Let's choose two movies: starwars, a sci-fi movie. And Liar Liar, a comedy.
#Now let's grab the user ratings for those two movies
```

```python
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
#We can then use corrwith() method to get correlations between two pandas series
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
#Let's clean this by removing NaN values and using a DataFrame instead of a series
corr_starwars = pd.DataFrame(similar_to_starwars,columns=['Correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

Now if we sort the dataframe by correlation, we should get the most similar movies, however note that we get some results that don't really make sense. This is because there are a lot of movies only watched once by users who also watched star wars (it was the most popular movie).

```python
corr_starwars.sort_values('Correlation',ascending=False).head(10)
```

```python
#Let's join this with Number of Ratings column in ratings data frame
corr_starwars = corr_starwars.join(ratings['ratings_count'])
corr_starwars.head()
```

```python
#et's filter out movies that have less than 100 reviews and Now sort the values and notice how the titles make a lot more sense
corr_starwars[corr_starwars['ratings_count']>100].sort_values('Correlation',ascending=False).head()
```

```python
#Now the same for the comedy Liar Liar
```

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['Correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['ratings_count'])
corr_liarliar[corr_liarliar['ratings_count']>100].sort_values('Correlation',ascending=False).head()
```
